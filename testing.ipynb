```python
import pandas as pd
df=pd.read_csv('GUIDE_Test.CSV')
df
```

```python
import pickle
with open('model.pkl', 'rb') as f:
    rfc=pickle.load(f)
```

```python
df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['MitreTechniques','ActionGrouped','ActionGranular','EmailClusterId','ThreatFamily','ResourceType','Roles','AntispamDirection','SuspicionLevel','LastVerdict'], inplace=True)
```

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder

x =  df[['Category','EvidenceRole','EntityType']]

y= df['IncidentGrade']

for i in x.columns:
    x[i] = LabelEncoder().fit_transform(x[i])
y= LabelEncoder().fit_transform(y)

chi_scores = chi2(x,y)

p_values = pd.Series(chi_scores[0],index = x.columns)
p_values.sort_values(ascending = False , inplace = True)
print(p_values)

```

```python
p_values.plot.bar()
```

```python
from sklearn.feature_selection import SelectKBest

anova= df[['Id','OrgId','IncidentId','AlertId','DetectorId','AlertTitle','DeviceId','Sha256','IpAddress','Url','AccountSid','AccountUpn','AccountObjectId','AccountName','DeviceName','NetworkMessageId','RegistryKey','RegistryValueName','RegistryValueData','ApplicationId','ApplicationName','OAuthApplicationId','FileName','FolderPath','ResourceIdName','OSFamily','OSVersion','CountryCode','State','City']]

selector=SelectKBest(chi2,k=15)
x_train_selected=selector.fit_transform(anova,y)

selected_features=anova.columns[selector.get_support()]
print(f'selected features are {selected_features.tolist()}')
```

```python
import pandas as pd

selected_features=selected_features.tolist()
selected_features.append('Category')
selected_features.append('EntityType')
```

```python
df=df[selected_features]
df
```

```python
for i in ['EntityType','Category']:
    df[i]=LabelEncoder().fit_transform(df[i])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)  
```

```python
y_pred=rfc.fit(x_train,y_train)
```

```python
y_predict=rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

macro_f1 = f1_score(y_test, y_predict, average='macro')
precision = precision_score(y_test, y_predict, average='macro')
recall = recall_score(y_test, y_predict, average='macro')

print(f"Macro F1 Score: {macro_f1}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming y_val are your actual labels and y_pred are your predicted labels
cm = confusion_matrix(y_test, y_predict)

# Define the labels corresponding to 0, 1, 2
labels = ['BP', 'FP', 'TP']

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')
plt.show()
```
